# Choosing dependent and independent variable

```python
data = df.iloc[:,3:]
data.drop(['imdb_id'], axis = 1, inplace= True)
data.head()
```

```python
y = data.iloc[:,[5]]
y.head()
```

```python
X = data.drop(['Verdict'], axis = 1)
X.head()
```

# Getting dummy for categorical data

```python
# Getting dummy for dependent data
y = pd.get_dummies(y, drop_first= True)
y.head()
```

```python
# Getting dummy for 'Released_month' in X
month_map = {'Jan':1,'Feb':2,'Mar':3,'Apr':4,'May':5,'Jun':6,'Jul':7,'Aug':8,'Sep':9,'Oct':10,'Nov':11,'Dec':12}
X['Released_month'] = X['Released_month'].map(month_map)
X.head()
```

```python
# correlation heatmap
cor = X.corr()
sns.heatmap(cor, annot = True, linewidths= 0.5,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Train test split

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
scaler = StandardScaler()
```

```python
X_train,X_test,y_train, y_test = train_test_split(X, y, train_size= 0.8, random_state= 42)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns = X.columns)
X_test = pd.DataFrame(X_test, columns = X.columns)
X_train.head()
```

```python
X_test.head()
```

```python
X_test.shape, y_test.shape
```

```python
X_train.shape, y_train.shape
```

# Biulding Neural Network

```python
import tensorflow
```

```python
from tensorflow import keras
from keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.layers import Dropout

model = Sequential()
model.add(Dense(64, activation = "relu", input_dim = X_train.shape[1])) # Increased neurons in the first layer
model.add(Dropout(0.2)) # Added dropout layer
model.add(Dense(32, activation = "relu")) # Added a second hidden layer
model.add(Dense(y_train.shape[1], activation="softmax")) # Output layer with softmax
```

```python
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
model.compile(optimizer = Adam(learning_rate=0.1), loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
history = model.fit(X_train,y_train, epochs=300, validation_split=0.2)
```

```python
model.layers[0].get_weights()
```

```python
model.layers[1].get_weights()
```

```python
prob = model.predict(X_test)
```

```python
prob
```

```python
prob.shape
```

```python
y_pred = np.zeros_like(prob)
y_pred[np.arange(len(prob)), np.argmax(prob, axis=1)] = 1
```

```python
y_pred = y_pred.astype(bool)
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.legend(['train', 'validation'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['train', 'validation'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
KNN_classifier = KNeighborsClassifier()
```

```python
param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11, 13],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan'],
    'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
    'p': [1, 2]
}
```

```python
KNN_model = GridSearchCV(KNN_classifier, param_grid, cv=5)
KNN_model
```

```python
KNN_model.fit(X_train, y_train)
```

```python
KNN_model.best_params_
```

```python
KNN_model.best_score_
```

```python
y_pred_KNN = KNN_model.predict(X_test)
y_pred_KNN
```

```python
accuracy_score(y_test, y_pred_KNN)
```

```python
print(classification_report(y_test, y_pred_KNN))
```
